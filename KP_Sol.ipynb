Let us look at correlation between future retuns and signals.


```python
window_size = 100

Asset_Returns = 1.0 * (Asset[window_size:] - Asset[:-window_size]) / Asset[:-window_size]
Asset_Returns = np.append(np.zeros(window_size), Asset_Returns)

df["Asset_Returns"] = Asset_Returns
print(df)
```

```python
for i in range(1,9):
  sig = "Signal"+str(i)

  s1 = df["Asset_Returns"]
  s2 = df[sig]

  roll_corr = s1.rolling(window_size).corr(s2)[window_size:]

  plt.figure()
  plt.plot(roll_corr.to_numpy())
  plt.xlabel(sig)
  plt.ylabel("Returns")
  plt.show()

  print("Corr > 0.0 : ", 100.0*len(roll_corr[roll_corr > 0]) / len(roll_corr))
  print("Corr > 0.2 : ", 100.0*len(roll_corr[roll_corr > 0.2]) / len(roll_corr))
  print("Corr < -0.2 : ", 100.0*len(roll_corr[roll_corr < -0.2]) / len(roll_corr))
  print("Median Corr: ", np.median(roll_corr.to_numpy()))  
```

It is very clear that Signal3 and Signal4 are very strong predictor of future returns.

Let us see what happens if we fit a linear model for returns on Signal3 and Signal4

```python
from sklearn.linear_model import LinearRegression

OLS_model = LinearRegression()
df_win = df[window_size:] 

X_train = df_win[["Signal3", "Signal4"]].to_numpy()
Y_train = df_win["Asset_Returns"].to_numpy()

print(X_train)
print(Y_train)

OLS_model.fit( X_train, Y_train)
Y_predict = OLS_model.predict(X_train)

plt.plot(Y_train , "r-", label = "Actual Returns")
plt.plot(Y_predict, "b-", label = "Fitted Returns")

plt.legend()
plt.show()

err = np.mean(np.abs(Y_train - Y_predict))/np.mean(np.abs(Y_train))
print("Error:" , err)
```

It appears that Signal1 and Signal5 are also slightly negetively correlated with the returns. Let's include them and see what happens:

```python
df_win = df[window_size:] 

X_train = df_win[["Signal1", "Signal3", "Signal4",  "Signal5"]].to_numpy()
Y_train = df_win["Asset_Returns"].to_numpy()

print(X_train)
print(Y_train)

OLS_model.fit( X_train, Y_train)
Y_predict = OLS_model.predict(X_train)

plt.plot(Y_train , "r-", label = "Actual Returns")
plt.plot(Y_predict, "b-", label = "Fitted Returns")

plt.legend()
plt.show()

err = np.mean(np.abs(Y_train - Y_predict))/np.mean(np.abs(Y_train))
print("Error:" , err)
```
